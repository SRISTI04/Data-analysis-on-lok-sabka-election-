```python
import requests
from bs4 import BeautifulSoup

# URL of the website to scrape
url_link = 'https://results.eci.gov.in'

# Send an HTTP request to the URL
res = requests.get(url_link)

# Check if the request was successful (status code 200)
if res.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(res.content, 'html.parser')

    # Now you can extract specific information using BeautifulSoup methods
    # For example, find all links
    links = soup.find_all('a')

    # Process further based on the structure of the website and the specific data you need

    # Example: Print all links
    for link in links:
        print(link.get('href'))
else:
    print(f"Failed to retrieve data. Status code: {res.status_code}")
```

```python
import matplotlib.pyplot as plt

# Example data
labels = ['A', 'B', 'C', 'D']
values = [10, 30, 20, 40]

# Example bar chart
plt.figure(figsize=(8, 6))
plt.bar(labels, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Example Bar Chart')
plt.grid(True)
plt.show()

# Example pie chart
plt.figure(figsize=(8, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=['pink', 'purple', 'lightblue', 'lightyellow'])
plt.title('Example Pie Chart')
plt.show()

# Example line plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='--', color='orange', label='Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Example Line Plot')
plt.legend()
plt.grid(True)
plt.show()
```
